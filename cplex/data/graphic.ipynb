```python
import matplotlib.pyplot as plt
```

PROLOG


```python
x_axis = [4,5,6,7]
y_axis = [0.013, 0.106, 0.644, 7.083]

plt.plot(x_axis, y_axis)
plt.title('prolog')
plt.xlabel('matrix NxN')
plt.ylabel('time(s)')
plt.show()
```

CPLEX

```python
x_axis = [4,5,6,7]
y_axis = [0.06, 0.23, 0.13, 0.84]

plt.plot(x_axis, y_axis)
plt.title('cplex')
plt.xlabel('matrix NxN')
plt.ylabel('time(s)')
plt.show()
```

Prolog & Cplex

```python
x_axis = [4,5,6,7]
y_cplex = [0.06, 0.23, 0.13, 0.84]
y_prolog = [0.013, 0.106, 0.644, 7.083]

plt.plot(x_axis, y_cplex, label="cplex")
plt.plot(x_axis, y_prolog, label="prolog")
plt.title('Comparing two platform')
plt.xlabel('matrix NxN')
plt.ylabel('time(s)')
plt.legend()
plt.show()
```
